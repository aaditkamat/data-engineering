```python
class RetailSalesAnalyzer():
    def __init__(self, url) -> None:
        self.url = url

    def load_data(self) -> pd.DataFrame:
        self.df = pd.read_csv(self.url)
        return self.df
    
    def calculate_stats(self) -> None:
        self.total_sales_per_product = self.df.groupby('Product').sum('Sales').Sales
        self.best_selling_product = self.total_sales_per_product.sort_values(ascending=False).iloc[0]
        self.average_daily_sales = self.df.groupby('Date').mean('Sales').Sales
        print(f"Total sales per product:{self.total_sales_per_product}\nBest selling product:{self.best_selling_product}\nAverage daily sales: {self.average_daily_sales}")
    
    def create_visualizations(self):
        fig, ax = plt.subplots(nrows=2, figsize=(15, 15))
        ax[0].set_ylabel('Average daily sales')
        ax[1].set_ylabel('Total Sales')
        sns.lineplot(self.average_daily_sales, ax=ax[1])
        sns.barplot(self.total_sales_per_product, ax=ax[0])
```
